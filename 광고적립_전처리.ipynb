```python
import pandas as pd
```

```python
df = pd.read_csv('IVE_광고적립.csv')
df.head()
```

# pub_sub_rel_id 컬럼 삭제

```python
df.drop('pub_sub_rel_id', axis=1, inplace=True)
```

```python
df.head()
```

# 광고주 단가 1~4원 삭제 및 확인

```python
idx = df[(df['show_cost'] >= 1) & (df['show_cost'] <= 4)].index
df.drop(idx , inplace=True)
```

```python
df[(df['show_cost'] >= 1) & (df['show_cost'] <= 4)]
```

# 광고 단가 1~4원 삭제 및 확인

```python
idx = df[(df['adv_cost'] >= 1) & (df['adv_cost'] <= 4)].index
df.drop(idx , inplace=True)
```

```python
df[(df['adv_cost'] >= 1) & (df['adv_cost'] <= 4)]
```

# 참여일시, 참여완료일시 object -> datetime 변환

```python
print('참여일시 변환 전:',df.click_date.dtype)
print('참여완료일시 변환 전:',df.regdate.dtype)
df['click_date'] = pd.to_datetime(df['click_date'])
df['regdate'] = pd.to_datetime(df['regdate'])
print('참여일시 변환 후:',df.click_date.dtype)
print('참여완료일시 변환 후:',df.regdate.dtype)
```
